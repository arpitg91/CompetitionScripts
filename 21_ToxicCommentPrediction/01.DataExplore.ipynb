```python
train=pd.read_csv('01.RawData/train.csv')
test=pd.read_csv('01.RawData/test.csv')
```

```python
tgts=['toxic','severe_toxic','obscene','threat','insult','identity_hate']
train['target']=train[tgts].max(axis=1)
train[tgts+['target']].mean()
```

```python
toxic=train[train['target']>0]
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS

mpl.rcParams['font.size']=12                #10 
mpl.rcParams['savefig.dpi']=100             #72 
mpl.rcParams['figure.subplot.bottom']=.1 

stopwords = set(STOPWORDS)

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(toxic['comment_text']))

print(wordcloud)
fig = plt.figure(1,figsize=(40,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
