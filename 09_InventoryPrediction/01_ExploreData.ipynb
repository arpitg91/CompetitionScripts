## Get EDD's of input DataSets 

```python
import pandas as pd

inputDir = '/data/arpit.goel/18_InventoryPrediction/02.ExtractedData/'
outputDir = '/data/arpit.goel/18_InventoryPrediction/03.DataSummary/'

for file in ['cliente_tabla.csv','sample_submission.csv','town_state.csv','producto_tabla.csv','test.csv','train.csv']:
#for file in ['cliente_tabla.csv']:
    df1 = pd.read_csv(inputDir+file,dtype=str)
    df2 = pd.read_csv(inputDir+file)
    edd1 = df1.describe().T
    edd2 = df2.describe().T
    if 'mean' in  edd2.columns.tolist():
        edd = pd.merge(edd1,edd2[edd2.columns.tolist()[1:]],left_index=True,right_index=True,how='left')
    else:
        edd = edd1
    edd.to_csv(outputDir+file)
```
